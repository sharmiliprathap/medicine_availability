# Handling Missing Values

```python
# Counting missing values

scraped_df.isnull().sum()
```

```python
# Dropping missing rows

scraped_df = scraped_df.dropna(subset=["strength", "salt_composition", "manufacturer"])
scraped_df.isnull().sum()
```

# Handling Duplicates

```python
# Counting duplicates

duplicated_count = scraped_df.duplicated(subset=["brand_name", "strength", "price"]).sum()
print(duplicated_count)
```

```python
# Create a view of all rows that match on Brand, Strength, and Price
duplicates_to_check = scraped_df[scraped_df.duplicated(subset=["brand_name", "strength", "price"], keep=False)]

# Sort them so you can see the "twins" together
duplicates_to_check = duplicates_to_check.sort_values(by=["brand_name", "price"])

# Look at the top 10 rows
duplicates_to_check.head(40)
```

```python
#drop duplicates

scraped_df = scraped_df.drop_duplicates(subset=["brand_name", "strength", "price"])
```

```python
scraped_df.duplicated(subset=["brand_name", "strength", "price"]).sum()
```

```python
scraped_df.shape
```

## Removing Combination drugs

```python
scraped_df = scraped_df[
    ~scraped_df['salt_composition'].str.contains(r'\+|,|/', regex=True, na=False)
]
```

```python
scraped_df.shape
```

## Extracting only primary molecule from salt

```python
import re

def extract_molecule(salt):
    match = re.match(r'([A-Za-z\s]+)', str(salt))
    return match.group(1).strip() if match else np.nan

scraped_df['Primary_Salt'] = scraped_df['salt_composition'].apply(extract_molecule)
```

```python
scraped_df.tail(50)
```

```python
scraped_df = scraped_df[
    scraped_df['drug_name'].str.strip() ==
    scraped_df['Primary_Salt'].str.strip()
]
```

```python
scraped_df.dtypes
```

# Standardizing Strength

```python
# Convert strength to numeric

scraped_df["strength"] = pd.to_numeric(scraped_df["strength"], errors="coerce")
master_df["strength"] = master_df["strength"].str.extract(r"(\d+\.?\d*)")
master_df["strength"] = pd.to_numeric(master_df["strength"], errors="coerce")
```

```python
# Delete strengths not in the master list by joining with master_df

df_filtered = pd.merge(
    scraped_df, 
    master_df, 
    left_on=['drug_name', 'strength'], 
    right_on=['medicine_name', 'strength'], 
    how='inner'
)

print(f"Rows before filtering: {len(scraped_df)}")
print(f"Rows after keeping only master strengths: {len(df_filtered)}")
```

```python
df_filtered.head(50)
```

```python
df_filtered['drug_name'].nunique()
```

```python
(df_filtered['drug_name'] == df_filtered['Primary_Salt']).all()
```

```python
# Drop the redundant column from the join
clean_df = df_filtered.drop(columns=['medicine_name', 'Primary_Salt', 'website'])

```

```python
clean_df.head()
```

# Cleaning Price

```python
# Cleaning the price column
clean_df["price"] = clean_df["price"].replace("", "", regex=True)
clean_df["price"] = pd.to_numeric(clean_df["price"], errors="coerce")
clean_df.head()
```

# Standardizing Availability

```python
# Standardizing availability

clean_df = clean_df[clean_df["availability"].isin(["In Stock", "Out of Stock"])]
```

```python
clean_df.shape
```

```python
clean_df["is_available"] = clean_df["availability"].apply(
    lambda x: 1 if x == "In Stock" else 0
)
```

```python
clean_df[["availability", "is_available"]].value_counts()
```

```python
clean_df["brand_name"] = clean_df["brand_name"].str.strip()
```

# Rename Columns

```python
# Rename columns if you want them to look professional for Power BI
clean_df = clean_df.rename(columns={
    'drug_name': 'Generic_Name',
    'strength': 'Strength',
    'brand_name': 'Brand_Name',
    'salt_composition': 'Molecule',
    'manufacturer': 'Manufacturer',
    'price': 'Price_INR',
    'availability': 'Availability',
    'is_available': 'Is_Available',
    'category': 'Therapeutic_Class'
})
```
